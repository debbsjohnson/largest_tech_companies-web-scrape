```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue")
soup = BeautifulSoup(page.text, "html")
```

```python
print(soup)
```

```python
soup.find("table")
```

```python
soup.find_all("table")[1]
```

```python
soup.find("table", class_="wikitable sortable plainrowheads")
```

```python
table = soup.find_all("table")[1]
print(table)
```

```python
world_titles = table.find_all("th")
world_titles
```

```python
titles = [title.text.strip() for title in world_titles]
titles
```

```python
df = pd.DataFrame(columns = titles)
df
```

```python
data = table.find_all("tr")
data
```

```python
for row in data:
    row_data = row.find_all("td")
    individual_row_data = [datas.text.strip() for datas in row_data]
    print(individual_row_data)
```

```python
for row in data[1:]:
    row_data = row.find_all("td")
    individual_row_data = [datas.text.strip() for datas in row_data]
    cleaned_row_data = [item for item in individual_row_data if item != '']
    
#     print(individual_row_data)

    if len(cleaned_row_data) == len(df.columns):
        length = len(df)
        df.loc[length] = cleaned_row_data
    else:
        print("row data has a different number of columns than the DataFrame")
    
```
